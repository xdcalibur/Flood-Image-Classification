```python
import os
import zipfile
import random
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import transforms, models
from PIL import Image, ImageFilter
from pathlib import Path
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from tqdm.auto import tqdm
import cv2

# Try to import google.colab for Drive mounting and display
try:
    from google.colab import drive
    from IPython.display import display
except ImportError:
    # Mock display for local environments
    def display(obj):
        print(obj)

# ==========================================
# 1. CONFIGURATION & SETUP
# ==========================================

BATCH_SIZE = 32
LEARNING_RATE = 1e-4
EPOCHS = 35
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42

def set_seed(seed=42):
    """Sets seed for reproducibility across numpy, random, and torch."""
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(SEED)
print(f"‚úÖ Using device: {DEVICE}")

# ==========================================
# 2. DATASET CLASS & TRANSFORMS
# ==========================================

class FloodDataset(Dataset):
    """
    Custom Dataset for loading Flood vs Non-Flood images.
    Expects a root directory structure or specific directories for classes.
    """
    def __init__(self, flood_dir, non_flood_dir, transform=None):
        self.flood_paths = [p for p in Path(flood_dir).rglob("*.*") if p.suffix.lower() in ['.jpg', '.jpeg', '.png']]
        self.non_flood_paths = [p for p in Path(non_flood_dir).rglob("*.*") if p.suffix.lower() in ['.jpg', '.jpeg', '.png']]

        self.samples = []
        for p in self.flood_paths:
            self.samples.append((p, 1)) # Label 1 for Flood
        for p in self.non_flood_paths:
            self.samples.append((p, 0)) # Label 0 for Non-Flood

        self.transform = transform

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        try:
            img = Image.open(path).convert("RGB")
        except Exception as e:
            print(f"Error loading {path}: {e}")
            # Return a black image in case of corruption to prevent crash
            img = Image.new('RGB', (256, 256))

        if self.transform:
            img = self.transform(img)

        return img, label

class BicubicDownUp:
    """
    Simulates low-resolution CCTV footage by downscaling and then upscaling
    using bicubic interpolation.
    """
    def __init__(self, factor=4, final_size=256):
        self.factor = factor
        self.final_size = final_size

    def __call__(self, img):
        img = img.resize((self.final_size, self.final_size), Image.BICUBIC)
        small_dim = (self.final_size // self.factor, self.final_size // self.factor)
        img_small = img.resize(small_dim, Image.BICUBIC)
        return img_small.resize((self.final_size, self.final_size), Image.BICUBIC)

class SimpleSR:
    """
    Simulates a basic Super-Resolution attempt (upscaling + sharpening).
    """
    def __init__(self, factor=4, final_size=256):
        self.factor = factor
        self.final_size = final_size

    def __call__(self, img):
        img = img.resize((self.final_size, self.final_size), Image.BICUBIC)
        small_dim = (self.final_size // self.factor, self.final_size // self.factor)
        img_small = img.resize(small_dim, Image.BICUBIC)
        img_up = img_small.resize((self.final_size, self.final_size), Image.BICUBIC)
        return img_up.filter(ImageFilter.UnsharpMask(radius=2, percent=150, threshold=3))

class TransformedSubset(Dataset):
    """
    Wrapper to apply specific transforms to a Subset of the data.
    Allows different transforms for Train and Val/Test splits.
    """
    def __init__(self, subset, transform):
        self.subset = subset
        self.transform = transform

    def __len__(self):
        return len(self.subset)

    def __getitem__(self, idx):
        original_img, label = self.subset[idx]
        if self.transform:
            img = self.transform(original_img)
        return img, label

# ==========================================
# 3. MODEL & TRAINING UTILITIES
# ==========================================

def get_model():
    """
    Tier 2 Alternative: Uses ResNet18 instead of EfficientNet_B0.
    """
    # 1. Load Pre-trained ResNet18
    model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

    # 2. Modify the classifier head for 2 classes (Flood vs Non-Flood)
    # ResNet uses 'fc' (Fully Connected) as the final layer
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, 2)

    return model.to(DEVICE)

def calculate_metrics(preds, labels):
    cm = confusion_matrix(labels, preds, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()
    eps = 1e-8
    accuracy = (tp + tn) / (tp + tn + fp + fn + eps)
    precision = tp / (tp + fp + eps)
    recall = tp / (tp + fn + eps)
    f1 = 2 * precision * recall / (precision + recall + eps)
    return {"acc": accuracy, "prec": precision, "rec": recall, "f1": f1, "cm": cm}

def evaluate(model, loader, criterion):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    epoch_loss = running_loss / len(loader.dataset)
    metrics = calculate_metrics(np.array(all_preds), np.array(all_labels))
    metrics['loss'] = epoch_loss
    return metrics

def run_training_experiment(experiment_name, full_dataset, indices, train_tf, val_tf, batch_size=32, epochs=10):
    print(f"\n{'='*20} Running Experiment: {experiment_name} {'='*20}")

    # Split logic (80/10/10)
    n_train = int(0.8 * len(indices))
    n_val = int(0.1 * len(indices))

    train_idx = indices[:n_train]
    val_idx = indices[n_train : n_train + n_val]
    test_idx = indices[n_train + n_val :]

    raw_train_subset = Subset(full_dataset, train_idx)
    raw_val_subset   = Subset(full_dataset, val_idx)
    raw_test_subset  = Subset(full_dataset, test_idx)

    train_ds = TransformedSubset(raw_train_subset, train_tf)
    val_ds   = TransformedSubset(raw_val_subset, val_tf)
    test_ds  = TransformedSubset(raw_test_subset, val_tf)

    # Note: num_workers=0 for safety in interactive environments
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, num_workers=0, pin_memory=True)
    val_loader   = DataLoader(val_ds, batch_size=batch_size, shuffle=False, num_workers=0, pin_memory=True)
    test_loader  = DataLoader(test_ds, batch_size=batch_size, shuffle=False, num_workers=0, pin_memory=True)

    model = get_model()
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

    best_f1 = 0.0
    history = {'train_loss': [], 'val_loss': [], 'val_f1': []}
    best_model_wts = model.state_dict().copy()

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=True)
        for inputs, labels in pbar:
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * inputs.size(0)

            pbar.set_postfix({'loss': loss.item()})

        epoch_train_loss = running_loss / len(train_ds)
        val_metrics = evaluate(model, val_loader, criterion)

        history['train_loss'].append(epoch_train_loss)
        history['val_loss'].append(val_metrics['loss'])
        history['val_f1'].append(val_metrics['f1'])

        print(f"Epoch {epoch+1} | Train Loss: {epoch_train_loss:.4f} | Val Loss: {val_metrics['loss']:.4f} | Val F1: {val_metrics['f1']:.4f}")

        if val_metrics['f1'] > best_f1:
            best_f1 = val_metrics['f1']
            best_model_wts = model.state_dict().copy()

    print(f"üéâ Best Validation F1: {best_f1:.4f}")
    model.load_state_dict(best_model_wts)

    # --- EVALUATE ON TEST SET BEFORE SAVING ---
    test_metrics = evaluate(model, test_loader, criterion)
    print(f"\n--- Test Results ({experiment_name}) ---")
    print(f"Accuracy:  {test_metrics['acc']:.4f}")
    print(f"F1 Score:  {test_metrics['f1']:.4f}")

    # --- CREATE SAVE PAYLOAD (Model + History + Metrics) ---
    save_payload = {
        'model_state_dict': model.state_dict(),
        'history': history,
        'test_metrics': test_metrics,
        'experiment_name': experiment_name,
        'architecture': 'ResNet18'
    }

    # Save models explicitly based on experiment name (UNIQUE RESNET NAMES)
    save_filename = None
    if "Baseline" in experiment_name:
        save_filename = "model_resnet_baseline.pth"
    elif "LowRes" in experiment_name:
        save_filename = "model_resnet_lowres.pth"
    elif "SuperRes" in experiment_name:
        save_filename = "model_resnet_superres.pth"

    if save_filename:
        torch.save(save_payload, save_filename)
        print(f"üíæ Model, History, and Metrics saved to {save_filename}")

    return history, test_metrics, model

# ==========================================
# 4. VISUALIZATION HELPERS
# ==========================================

def visualize_data_qualities(image_path):
    path_obj = Path(image_path)
    if not path_obj.exists():
        print(f"‚ùå Cannot visualize data quality: Image not found at {image_path}")
        return

    try:
        img_pil = Image.open(path_obj).convert("RGB")
    except Exception as e:
        print(f"‚ùå Error loading image for visualization: {e}")
        return

    t_bicubic = BicubicDownUp(factor=4, final_size=256)
    t_sr = SimpleSR(factor=4, final_size=256)
    t_resize = transforms.Resize((256, 256))

    img_orig = t_resize(img_pil)
    img_low = t_bicubic(img_pil)
    img_sr = t_sr(img_pil)

    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(img_orig)
    plt.title(f"Original (Resized)\nSize: {img_orig.size}")
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(img_low)
    plt.title("Low Res Simulation\n(Bicubic Down->Up)")
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(img_sr)
    plt.title("Simple Super Res\n(Sharpening)")
    plt.axis('off')

    plt.suptitle(f"Data Quality Comparison: {os.path.basename(image_path)}", fontsize=16)
    plt.tight_layout()
    plt.show()

# ==========================================
# 5. MAIN TRAINING EXECUTION
# ==========================================

if __name__ == "__main__":
    # --- 1. SETUP & EXTRACTION ---
    print("üöÄ Starting Setup (ResNet Edition)...")

    # Original Paths from user request
    shortcut_path = Path('/content/drive/MyDrive/Scientific ML Project')
    project_folder = shortcut_path / "New Final Project"
    dataset_folder = project_folder / "Dataset"

    zip_path = dataset_folder / "flood_data_alley.zip"
    extract_path = Path('/content/dataset')

    # Try to import google.colab to see if we are in colab, if so mount drive
    try:
        from google.colab import drive
        if not os.path.exists('/content/drive'):
            drive.mount('/content/drive', force_remount=True)
    except:
        pass

    if extract_path.exists():
        print("‚úÖ Data directory exists.")
    elif zip_path.exists():
        print(f"üìÇ Extracting zip...")
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        print("‚úÖ Extraction complete.")
    else:
        print("‚ö†Ô∏è Zip file not found. Ensure data is placed in './dataset' or provide zip.")

    # --- 2. DATA PREP (ROBUST DIRECTORY FINDING) ---
    base_dir = Path(extract_path)
    print(f"üìÇ Searching for data folders in {base_dir}...")

    all_subdirs = [p for p in base_dir.rglob("*") if p.is_dir()]

    flood_dir = None
    non_flood_dir = None

    for d in all_subdirs:
        name = d.name.lower()
        if name == "flood images low res":
            flood_dir = d
        elif name == "non flood images":
            non_flood_dir = d

    if not flood_dir:
        for d in all_subdirs:
            if "flood" in d.name.lower() and "non" not in d.name.lower():
                flood_dir = d
                break

    if not non_flood_dir:
        for d in all_subdirs:
            if "non" in d.name.lower() and "flood" in d.name.lower():
                non_flood_dir = d
                break

    # --- 3. TRAINING CONFIG ---
    # Options: "Baseline_Original", "LowRes_Bicubic", "SuperRes_Simple"
    EXPERIMENTS_TO_RUN = ["Baseline_Original", "LowRes_Bicubic", "SuperRes_Simple"]

    if flood_dir and non_flood_dir and EXPERIMENTS_TO_RUN:
        print(f"‚úÖ Found Flood Dir: {flood_dir}")
        print(f"‚úÖ Found Non-Flood Dir: {non_flood_dir}")

        full_dataset = FloodDataset(flood_dir, non_flood_dir)

        # --- NEW: PRINT DATASET COUNTS ---
        print(f"üìä Dataset Statistics:")
        print(f"   Total Images: {len(full_dataset)}")
        print(f"   - Flood Images: {len(full_dataset.flood_paths)}")
        print(f"   - Non-Flood Images: {len(full_dataset.non_flood_paths)}")

        indices = list(range(len(full_dataset)))
        random.shuffle(indices)

        # --- A. Visualize Transforms ---
        print("\nüñºÔ∏è Visualizing Data Quality methods...")
        if len(full_dataset.samples) > 0:
            sample_img_path, _ = full_dataset.samples[0]
            visualize_data_qualities(sample_img_path)

        # --- B. Define Experiments (Augmentations) ---
        tf_standard_train = transforms.Compose([
            transforms.Resize((256, 256)),
            transforms.RandomHorizontalFlip(),
            transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1),
            transforms.RandomGrayscale(p=0.1),
            transforms.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5)),
            transforms.RandomRotation(degrees=15),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        tf_standard_eval = transforms.Compose([
            transforms.Resize((256, 256)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

        tf_bicubic_train = transforms.Compose([
            BicubicDownUp(factor=4, final_size=256),
            transforms.RandomHorizontalFlip(),
            transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1),
            transforms.RandomGrayscale(p=0.1),
            transforms.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5)),
            transforms.RandomRotation(degrees=15),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        tf_bicubic_eval = transforms.Compose([
            BicubicDownUp(factor=4, final_size=256),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

        tf_sr_train = transforms.Compose([
            SimpleSR(factor=4, final_size=256),
            transforms.RandomHorizontalFlip(),
            transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1),
            transforms.RandomGrayscale(p=0.1),
            transforms.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5)),
            transforms.RandomRotation(degrees=15),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        tf_sr_eval = transforms.Compose([
            SimpleSR(factor=4, final_size=256),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

        # Definition of all possible experiments
        all_experiments_map = {
            "Baseline_Original": (tf_standard_train, tf_standard_eval),
            "LowRes_Bicubic":    (tf_bicubic_train, tf_bicubic_eval),
            "SuperRes_Simple":   (tf_sr_train, tf_sr_eval)
        }

        # Filter based on user selection
        experiments = {k: v for k, v in all_experiments_map.items() if k in EXPERIMENTS_TO_RUN}

        if not experiments:
            print("‚ö†Ô∏è No experiments selected for training.")
        else:
            # --- C. Run Training ---
            print("\nüöÄ Starting Multi-Experiment Training (ResNet)...")
            results = {}

            for name, (t_train, t_eval) in experiments.items():
                hist, metrics, model = run_training_experiment(
                    name,
                    full_dataset,
                    indices,
                    t_train,
                    t_eval,
                    epochs=EPOCHS
                )
                results[name] = {'history': hist, 'metrics': metrics}

            # --- D. Evaluation Metrics Display ---
            print("\nüìä === Experiment Performance Comparison ===")
            metrics_data = []
            for name, data in results.items():
                m = data['metrics']
                metrics_data.append({
                    "Experiment": name,
                    "Accuracy": m['acc'],
                    "Precision": m['prec'],
                    "Recall": m['rec'],
                    "F1 Score": m['f1']
                })

            df_metrics = pd.DataFrame(metrics_data)
            print(df_metrics)

            # Plot Confusion Matrix
            plt.figure(figsize=(18, 5))
            for i, (name, data) in enumerate(results.items()):
                plt.subplot(1, 3, i+1)
                cm = data['metrics']['cm']
                sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                            xticklabels=['Non-Flood', 'Flood'],
                            yticklabels=['Non-Flood', 'Flood'])
                plt.title(f"{name}\nF1: {data['metrics']['f1']:.3f}")
                plt.ylabel('True Label')
                plt.xlabel('Predicted Label')
            plt.tight_layout()
            plt.show()

    else:
        # Warning if paths are wrong or training skipped
        print("\n‚ùå TRAINING SKIPPED: Could not find dataset folders OR no experiments selected.")
        print(f"   Flood Dir detected: {flood_dir}")
        print(f"   Non-Flood Dir detected: {non_flood_dir}")
        print("   Please check your Drive paths or Zip extraction.")

    print("\n‚úÖ Training Script Complete.")
    print("‚¨áÔ∏è  SCROLL DOWN TO COPY THE STANDALONE INFERENCE CELL")
```

```python
# ==================================================================================
# ==================================================================================
#
#   7. STANDALONE INFERENCE CELL (COPY EVERYTHING BELOW TO A NEW COLAB CELL)
#      This section contains self-contained code to load a model and test images.
#
# ==================================================================================
# ==================================================================================

import os
import torch
import torch.nn as nn
from torchvision import transforms, models
from PIL import Image
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import cv2
import seaborn as sns
from sklearn.metrics import confusion_matrix, f1_score
from tqdm.auto import tqdm

# --- CONFIGURATION FOR STANDALONE TESTING ---
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# RESTORED ORIGINAL PATHS:
MODEL_LOAD_PATH = "model_resnet_baseline.pth"  # Defaulting to ResNet Baseline
TEST_IMAGE_PATH = "/content/drive/MyDrive/Scientific ML Project/New Final Project/Dataset/New photos/flood/48.jpg"

# UPDATED TEST PATHS:
TEST_FOLDER_FLOOD = "/content/drive/MyDrive/Scientific ML Project/New Final Project/Dataset/Alleyphoto_test/flood"
TEST_FOLDER_NON_FLOOD = "/content/drive/MyDrive/Scientific ML Project/New Final Project/Dataset/Alleyphoto_test/no_flood"

# --- MODEL DEFINITION (ResNet18) ---
def get_model():
    model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, 2)
    return model.to(DEVICE)

# --- UTILITIES ---
class GradCAM:
    """Updated for ResNet18 structure."""
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.handle_forward = target_layer.register_forward_hook(self.save_activation)
        self.handle_backward = target_layer.register_full_backward_hook(self.save_gradient)

    def save_activation(self, module, input, output):
        self.activation = output

    def save_gradient(self, module, grad_input, grad_output):
        self.gradients = grad_output[0]

    def remove_hooks(self):
        self.handle_forward.remove()
        self.handle_backward.remove()

    def __call__(self, x, class_idx=None):
        output = self.model(x)
        if class_idx is None:
            class_idx = torch.argmax(output, dim=1)
        self.model.zero_grad()
        target = output[0, class_idx]
        target.backward()

        gradients = self.gradients[0]
        pooled_gradients = torch.mean(gradients, dim=[1, 2])
        activations = self.activation[0]

        for i in range(activations.shape[0]):
            activations[i, :, :] *= pooled_gradients[i]

        heatmap = torch.mean(activations, dim=0).cpu().detach()
        heatmap = np.maximum(heatmap, 0)
        heatmap /= torch.max(heatmap) + 1e-8
        return heatmap.numpy()

def inverse_normalize(tensor):
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = tensor.cpu().numpy().transpose((1, 2, 0))
    img = std * img + mean
    img = np.clip(img, 0, 1)
    return img

def predict_single_image(model, image_path, device='cuda'):
    path_obj = Path(image_path)
    if not path_obj.exists():
        print(f"‚ùå Error: Image not found at {image_path}")
        return

    transform = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    try:
        img_pil = Image.open(path_obj).convert("RGB")
    except Exception as e:
        print(f"‚ùå Could not open image: {e}")
        return

    input_tensor = transform(img_pil).unsqueeze(0).to(device)
    model.eval()

    # Target ResNet's last feature block (layer4)
    target_layer = model.layer4[-1]
    grad_cam = GradCAM(model, target_layer)

    output = model(input_tensor)
    probs = torch.nn.functional.softmax(output, dim=1)
    confidence, predicted_class = torch.max(probs, 1)

    confidence_score = confidence.item()
    pred_idx = predicted_class.item()
    class_names = {0: "Non-Flood", 1: "Flood"}
    pred_label = class_names.get(pred_idx, "Unknown")

    heatmap = grad_cam(input_tensor, class_idx=pred_idx)
    grad_cam.remove_hooks()

    img_np = inverse_normalize(input_tensor.squeeze(0))
    heatmap = cv2.resize(heatmap, (img_np.shape[1], img_np.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)

    superimposed = np.float32(heatmap) / 255 + img_np
    superimposed = superimposed / np.max(superimposed)

    print(f"\nüì∏ Image: {os.path.basename(image_path)}")
    print(f"ü§ñ Prediction: {pred_label}")
    print(f"üìä Confidence: {confidence_score:.4f} ({confidence_score*100:.2f}%)")

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img_np)
    plt.title(f"Prediction: {pred_label} ({confidence_score:.1%})")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(superimposed)
    plt.title("Grad-CAM Attention (ResNet)")
    plt.axis('off')
    plt.show()

def run_batch_test(model, folder_path, true_label_name, device=DEVICE):
    folder_path = Path(folder_path)
    if not folder_path.exists():
        print(f"‚ö†Ô∏è Folder not found or skipped: {folder_path}")
        return

    print(f"\nüìÇ Testing images in: {folder_path}")
    image_files = [f for f in folder_path.iterdir() if f.suffix.lower() in ('.jpg', '.jpeg', '.png')]
    total_count = len(image_files)

    if total_count == 0:
        print("   No images found.")
        return

    # Accumulate predictions for metrics
    all_preds = []
    all_labels = []
    # 1 for Flood, 0 for Non-Flood
    target_label = 1 if true_label_name == "Flood" else 0

    transform = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    print(f"   Found {total_count} images. Processing...")
    correct_count = 0

    for img_path in tqdm(image_files, leave=True, desc=f"Testing {true_label_name}"):
        try:
            img_pil = Image.open(img_path).convert("RGB")
            input_tensor = transform(img_pil).unsqueeze(0).to(device)

            model.eval()
            with torch.no_grad():
                output = model(input_tensor)
                probs = torch.nn.functional.softmax(output, dim=1)
                confidence, predicted_idx = torch.max(probs, 1)

            pred_idx = predicted_idx.item()
            all_preds.append(pred_idx)
            all_labels.append(target_label)

            if pred_idx == target_label:
                correct_count += 1

        except Exception as e:
            print(f"   Error processing {img_path.name}: {e}")

    # Results and Metrics
    accuracy = (correct_count / total_count) * 100
    f1 = f1_score(all_labels, all_preds, labels=[0, 1], average='binary', zero_division=0)
    cm = confusion_matrix(all_labels, all_preds, labels=[0, 1])

    print(f"   üëâ Results for {true_label_name} folder:")
    print(f"      Accuracy: {correct_count}/{total_count} ({accuracy:.2f}%)")
    print(f"      F1 Score: {f1:.4f}")

    # Plot Simple Confusion Matrix
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Non-Flood', 'Flood'],
                yticklabels=['Non-Flood', 'Flood'])
    plt.title(f"CM: {true_label_name} Test")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

# --- MAIN INFERENCE EXECUTION ---
if __name__ == "__main__":
    # Ensure Drive is mounted
    try:
        from google.colab import drive
        if not os.path.exists('/content/drive'):
            drive.mount('/content/drive', force_remount=True)
    except:
        pass

    if os.path.exists(MODEL_LOAD_PATH):
        print(f"üîÑ Loading checkpoint from {MODEL_LOAD_PATH}...")

        # Load the raw checkpoint
        checkpoint = torch.load(MODEL_LOAD_PATH, map_location=DEVICE, weights_only=False)

        model = get_model()

        # Check if it's the new dictionary format or old state_dict
        if isinstance(checkpoint, dict) and 'model_state_dict' in checkpoint:
            print("   ‚úÖ Full checkpoint dictionary detected!")
            model.load_state_dict(checkpoint['model_state_dict'])

            # Print Stored Metrics for Easy Access
            if 'test_metrics' in checkpoint:
                tm = checkpoint['test_metrics']
                print("\nüìä --- Stored Test Metrics (from training) ---")
                print(f"   Accuracy:  {tm.get('acc', 'N/A'):.4f}")
                print(f"   Precision: {tm.get('prec', 'N/A'):.4f}")
                print(f"   Recall:    {tm.get('rec', 'N/A'):.4f}")
                print(f"   F1 Score:  {tm.get('f1', 'N/A'):.4f}")

            if 'history' in checkpoint:
                print("\nüìâ --- Training History Available ---")
                hist = checkpoint['history']
                # Quick plot of history if available
                plt.figure(figsize=(10, 4))
                plt.subplot(1, 2, 1)
                plt.plot(hist['train_loss'], label='Train Loss')
                plt.plot(hist['val_loss'], label='Val Loss')
                plt.title("Loss Curve")
                plt.legend()

                plt.subplot(1, 2, 2)
                plt.plot(hist['val_f1'], label='Val F1', color='orange')
                plt.title("F1 Score Curve")
                plt.legend()
                plt.show()

        else:
            print("   ‚ö†Ô∏è Legacy state_dict detected (Weights only). Loading weights...")
            model.load_state_dict(checkpoint)

        # 1. Test Single Image
        if TEST_IMAGE_PATH:
            predict_single_image(model, TEST_IMAGE_PATH, device=DEVICE)

        # 2. Test Folders
        if TEST_FOLDER_FLOOD:
            run_batch_test(model, TEST_FOLDER_FLOOD, "Flood", device=DEVICE)

        if TEST_FOLDER_NON_FLOOD:
            run_batch_test(model, TEST_FOLDER_NON_FLOOD, "Non-Flood", device=DEVICE)

    else:
        print(f"‚ùå Model file '{MODEL_LOAD_PATH}' not found. Please run training first.")
```

```python
import os
import zipfile
import random
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import transforms, models
from PIL import Image, ImageFilter
from pathlib import Path
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from tqdm.auto import tqdm
import cv2

# Try to import google.colab for Drive mounting and display
try:
    from google.colab import drive
    from IPython.display import display
except ImportError:
    # Mock display for local environments
    def display(obj):
        print(obj)

# ==========================================
# 1. CONFIGURATION & SETUP
# ==========================================

BATCH_SIZE = 32
LEARNING_RATE = 1e-4
EPOCHS = 35
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42

def set_seed(seed=42):
    """Sets seed for reproducibility across numpy, random, and torch."""
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(SEED)
print(f"‚úÖ Using device: {DEVICE}")

# ==========================================
# 2. DATASET CLASS & TRANSFORMS
# ==========================================

class FloodDataset(Dataset):
    """
    Custom Dataset for loading Flood vs Non-Flood images.
    Expects a root directory structure or specific directories for classes.
    """
    def __init__(self, flood_dir, non_flood_dir, transform=None):
        self.flood_paths = [p for p in Path(flood_dir).rglob("*.*") if p.suffix.lower() in ['.jpg', '.jpeg', '.png']]
        self.non_flood_paths = [p for p in Path(non_flood_dir).rglob("*.*") if p.suffix.lower() in ['.jpg', '.jpeg', '.png']]

        self.samples = []
        for p in self.flood_paths:
            self.samples.append((p, 1)) # Label 1 for Flood
        for p in self.non_flood_paths:
            self.samples.append((p, 0)) # Label 0 for Non-Flood

        self.transform = transform

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        try:
            img = Image.open(path).convert("RGB")
        except Exception as e:
            print(f"Error loading {path}: {e}")
            # Return a black image in case of corruption to prevent crash
            img = Image.new('RGB', (256, 256))

        if self.transform:
            img = self.transform(img)

        return img, label

class BicubicDownUp:
    """
    Simulates low-resolution CCTV footage by downscaling and then upscaling
    using bicubic interpolation.
    """
    def __init__(self, factor=4, final_size=256):
        self.factor = factor
        self.final_size = final_size

    def __call__(self, img):
        img = img.resize((self.final_size, self.final_size), Image.BICUBIC)
        small_dim = (self.final_size // self.factor, self.final_size // self.factor)
        img_small = img.resize(small_dim, Image.BICUBIC)
        return img_small.resize((self.final_size, self.final_size), Image.BICUBIC)

class SimpleSR:
    """
    Simulates a basic Super-Resolution attempt (upscaling + sharpening).
    """
    def __init__(self, factor=4, final_size=256):
        self.factor = factor
        self.final_size = final_size

    def __call__(self, img):
        img = img.resize((self.final_size, self.final_size), Image.BICUBIC)
        small_dim = (self.final_size // self.factor, self.final_size // self.factor)
        img_small = img.resize(small_dim, Image.BICUBIC)
        img_up = img_small.resize((self.final_size, self.final_size), Image.BICUBIC)
        return img_up.filter(ImageFilter.UnsharpMask(radius=2, percent=150, threshold=3))

class TransformedSubset(Dataset):
    """
    Wrapper to apply specific transforms to a Subset of the data.
    Allows different transforms for Train and Val/Test splits.
    """
    def __init__(self, subset, transform):
        self.subset = subset
        self.transform = transform

    def __len__(self):
        return len(self.subset)

    def __getitem__(self, idx):
        original_img, label = self.subset[idx]
        if self.transform:
            img = self.transform(original_img)
        return img, label

# ==========================================
# 3. MODEL & TRAINING UTILITIES
# ==========================================

def get_model():
    """
    Tier 1 Upgrade: Uses EfficientNet_B0 instead of ResNet18.
    """
    # Load Pre-trained EfficientNet_B0
    model = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)

    # Modify the classifier head for 2 classes (Flood vs Non-Flood)
    num_ftrs = model.classifier[1].in_features
    model.classifier[1] = nn.Linear(num_ftrs, 2)

    return model.to(DEVICE)

def calculate_metrics(preds, labels):
    cm = confusion_matrix(labels, preds, labels=[0, 1])
    tn, fp, fn, tp = cm.ravel()
    eps = 1e-8
    accuracy = (tp + tn) / (tp + tn + fp + fn + eps)
    precision = tp / (tp + fp + eps)
    recall = tp / (tp + fn + eps)
    f1 = 2 * precision * recall / (precision + recall + eps)
    return {"acc": accuracy, "prec": precision, "rec": recall, "f1": f1, "cm": cm}

def evaluate(model, loader, criterion):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    epoch_loss = running_loss / len(loader.dataset)
    metrics = calculate_metrics(np.array(all_preds), np.array(all_labels))
    metrics['loss'] = epoch_loss
    return metrics

def run_training_experiment(experiment_name, full_dataset, indices, train_tf, val_tf, batch_size=32, epochs=10):
    print(f"\n{'='*20} Running Experiment: {experiment_name} {'='*20}")

    # Split logic (80/10/10)
    n_train = int(0.8 * len(indices))
    n_val = int(0.1 * len(indices))

    train_idx = indices[:n_train]
    val_idx = indices[n_train : n_train + n_val]
    test_idx = indices[n_train + n_val :]

    raw_train_subset = Subset(full_dataset, train_idx)
    raw_val_subset   = Subset(full_dataset, val_idx)
    raw_test_subset  = Subset(full_dataset, test_idx)

    train_ds = TransformedSubset(raw_train_subset, train_tf)
    val_ds   = TransformedSubset(raw_val_subset, val_tf)
    test_ds  = TransformedSubset(raw_test_subset, val_tf)

    # Note: num_workers=0 for safety in interactive environments
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, num_workers=0, pin_memory=True)
    val_loader   = DataLoader(val_ds, batch_size=batch_size, shuffle=False, num_workers=0, pin_memory=True)
    test_loader  = DataLoader(test_ds, batch_size=batch_size, shuffle=False, num_workers=0, pin_memory=True)

    model = get_model()
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

    best_f1 = 0.0
    history = {'train_loss': [], 'val_loss': [], 'val_f1': []}
    best_model_wts = model.state_dict().copy()

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        # UPDATED: leave=True prevents the bar from disappearing
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=True)
        for inputs, labels in pbar:
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * inputs.size(0)

            # Optional: Update pbar with current loss
            pbar.set_postfix({'loss': loss.item()})

        epoch_train_loss = running_loss / len(train_ds)
        val_metrics = evaluate(model, val_loader, criterion)

        history['train_loss'].append(epoch_train_loss)
        history['val_loss'].append(val_metrics['loss'])
        history['val_f1'].append(val_metrics['f1'])

        print(f"Epoch {epoch+1} | Train Loss: {epoch_train_loss:.4f} | Val Loss: {val_metrics['loss']:.4f} | Val F1: {val_metrics['f1']:.4f}")

        if val_metrics['f1'] > best_f1:
            best_f1 = val_metrics['f1']
            best_model_wts = model.state_dict().copy()

    print(f"üéâ Best Validation F1: {best_f1:.4f}")
    model.load_state_dict(best_model_wts)

    # --- EVALUATE ON TEST SET BEFORE SAVING ---
    test_metrics = evaluate(model, test_loader, criterion)
    print(f"\n--- Test Results ({experiment_name}) ---")
    print(f"Accuracy:  {test_metrics['acc']:.4f}")
    print(f"F1 Score:  {test_metrics['f1']:.4f}")

    # --- CREATE SAVE PAYLOAD (Model + History + Metrics) ---
    save_payload = {
        'model_state_dict': model.state_dict(),
        'history': history,
        'test_metrics': test_metrics,
        'experiment_name': experiment_name,
        'architecture': 'EfficientNet_B0'
    }

    # Save models explicitly based on experiment name (UNIQUE EFFICIENTNET NAMES)
    # Using 'efficientnet' in filename to avoid conflict with resnet files
    save_filename = None
    if "Baseline" in experiment_name:
        save_filename = "model_efficientnet_baseline.pth"
    elif "LowRes" in experiment_name:
        save_filename = "model_efficientnet_lowres.pth"
    elif "SuperRes" in experiment_name:
        save_filename = "model_efficientnet_superres.pth"

    if save_filename:
        torch.save(save_payload, save_filename)
        print(f"üíæ Model, History, and Metrics saved to {save_filename}")

    return history, test_metrics, model

# ==========================================
# 4. VISUALIZATION HELPERS
# ==========================================

def visualize_data_qualities(image_path):
    path_obj = Path(image_path)
    if not path_obj.exists():
        print(f"‚ùå Cannot visualize data quality: Image not found at {image_path}")
        return

    try:
        img_pil = Image.open(path_obj).convert("RGB")
    except Exception as e:
        print(f"‚ùå Error loading image for visualization: {e}")
        return

    t_bicubic = BicubicDownUp(factor=4, final_size=256)
    t_sr = SimpleSR(factor=4, final_size=256)
    t_resize = transforms.Resize((256, 256))

    img_orig = t_resize(img_pil)
    img_low = t_bicubic(img_pil)
    img_sr = t_sr(img_pil)

    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(img_orig)
    plt.title(f"Original (Resized)\nSize: {img_orig.size}")
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(img_low)
    plt.title("Low Res Simulation\n(Bicubic Down->Up)")
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(img_sr)
    plt.title("Simple Super Res\n(Sharpening)")
    plt.axis('off')

    plt.suptitle(f"Data Quality Comparison: {os.path.basename(image_path)}", fontsize=16)
    plt.tight_layout()
    plt.show()

# ==========================================
# 5. MAIN TRAINING EXECUTION
# ==========================================

if __name__ == "__main__":
    # --- 1. SETUP & EXTRACTION ---
    print("üöÄ Starting Setup (EfficientNet Edition)...")

    # Original Paths from user request
    shortcut_path = Path('/content/drive/MyDrive/Scientific ML Project')
    project_folder = shortcut_path / "New Final Project"
    dataset_folder = project_folder / "Dataset"

    # UPDATED ZIP NAME
    zip_path = dataset_folder / "flood_data_alley.zip"
    extract_path = Path('/content/dataset')

    # Try to import google.colab to see if we are in colab, if so mount drive
    try:
        from google.colab import drive
        if not os.path.exists('/content/drive'):
            drive.mount('/content/drive', force_remount=True)
    except:
        pass

    if extract_path.exists():
        print("‚úÖ Data directory exists.")
    elif zip_path.exists():
        print(f"üìÇ Extracting zip...")
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        print("‚úÖ Extraction complete.")
    else:
        print("‚ö†Ô∏è Zip file not found. Ensure data is placed in './dataset' or provide zip.")

    # --- 2. DATA PREP (ROBUST DIRECTORY FINDING) ---
    base_dir = Path(extract_path)
    print(f"üìÇ Searching for data folders in {base_dir}...")

    # Helper: Find strict directory matches
    # This prevents 'rglob' from picking up files like 'flood_4856.jpg'
    all_subdirs = [p for p in base_dir.rglob("*") if p.is_dir()]

    flood_dir = None
    non_flood_dir = None

    # Strategy: Look for the specific names you provided
    for d in all_subdirs:
        # Case insensitive checks
        name = d.name.lower()
        if name == "flood images low res":
            flood_dir = d
        elif name == "non flood images":
            non_flood_dir = d

    # Fallback if exact names fail: strict substring search on directories only
    if not flood_dir:
        for d in all_subdirs:
            # Must have 'flood', must NOT have 'non'
            if "flood" in d.name.lower() and "non" not in d.name.lower():
                flood_dir = d
                break

    if not non_flood_dir:
        for d in all_subdirs:
            # Must have 'non' and 'flood'
            if "non" in d.name.lower() and "flood" in d.name.lower():
                non_flood_dir = d
                break

    # --- 3. TRAINING CONFIG ---
    # Options: "Baseline_Original", "LowRes_Bicubic", "SuperRes_Simple"
    EXPERIMENTS_TO_RUN = ["Baseline_Original", "LowRes_Bicubic", "SuperRes_Simple"]

    if flood_dir and non_flood_dir and EXPERIMENTS_TO_RUN:
        print(f"‚úÖ Found Flood Dir: {flood_dir}")
        print(f"‚úÖ Found Non-Flood Dir: {non_flood_dir}")

        full_dataset = FloodDataset(flood_dir, non_flood_dir)

        # --- NEW: PRINT DATASET COUNTS ---
        print(f"üìä Dataset Statistics:")
        print(f"   Total Images: {len(full_dataset)}")
        print(f"   - Flood Images: {len(full_dataset.flood_paths)}")
        print(f"   - Non-Flood Images: {len(full_dataset.non_flood_paths)}")

        indices = list(range(len(full_dataset)))
        random.shuffle(indices)

        # --- A. Visualize Transforms ---
        print("\nüñºÔ∏è Visualizing Data Quality methods...")
        if len(full_dataset.samples) > 0:
            sample_img_path, _ = full_dataset.samples[0]
            visualize_data_qualities(sample_img_path)

        # --- B. Define Experiments (Augmentations) ---
        tf_standard_train = transforms.Compose([
            transforms.Resize((256, 256)),
            transforms.RandomHorizontalFlip(),
            transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1),
            transforms.RandomGrayscale(p=0.1),
            transforms.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5)),
            transforms.RandomRotation(degrees=15),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        tf_standard_eval = transforms.Compose([
            transforms.Resize((256, 256)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

        tf_bicubic_train = transforms.Compose([
            BicubicDownUp(factor=4, final_size=256),
            transforms.RandomHorizontalFlip(),
            transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1),
            transforms.RandomGrayscale(p=0.1),
            transforms.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5)),
            transforms.RandomRotation(degrees=15), # Added for parity
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        tf_bicubic_eval = transforms.Compose([
            BicubicDownUp(factor=4, final_size=256),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

        tf_sr_train = transforms.Compose([
            SimpleSR(factor=4, final_size=256),
            transforms.RandomHorizontalFlip(),
            transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1),
            transforms.RandomGrayscale(p=0.1),
            transforms.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5)),
            transforms.RandomRotation(degrees=15), # Added for parity
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        tf_sr_eval = transforms.Compose([
            SimpleSR(factor=4, final_size=256),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

        # Definition of all possible experiments
        all_experiments_map = {
            "Baseline_Original": (tf_standard_train, tf_standard_eval),
            "LowRes_Bicubic":    (tf_bicubic_train, tf_bicubic_eval),
            "SuperRes_Simple":   (tf_sr_train, tf_sr_eval)
        }

        # Filter based on user selection
        experiments = {k: v for k, v in all_experiments_map.items() if k in EXPERIMENTS_TO_RUN}

        if not experiments:
            print("‚ö†Ô∏è No experiments selected for training.")
        else:
            # --- C. Run Training ---
            print("\nüöÄ Starting Multi-Experiment Training...")
            results = {}

            for name, (t_train, t_eval) in experiments.items():
                hist, metrics, model = run_training_experiment(
                    name,
                    full_dataset,
                    indices,
                    t_train,
                    t_eval,
                    epochs=EPOCHS
                )
                results[name] = {'history': hist, 'metrics': metrics}

            # --- D. Evaluation Metrics Display ---
            print("\nüìä === Experiment Performance Comparison ===")
            metrics_data = []
            for name, data in results.items():
                m = data['metrics']
                metrics_data.append({
                    "Experiment": name,
                    "Accuracy": m['acc'],
                    "Precision": m['prec'],
                    "Recall": m['rec'],
                    "F1 Score": m['f1']
                })

            df_metrics = pd.DataFrame(metrics_data)
            print(df_metrics)

            # Plot Confusion Matrix
            plt.figure(figsize=(18, 5))
            for i, (name, data) in enumerate(results.items()):
                plt.subplot(1, 3, i+1)
                cm = data['metrics']['cm']
                sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                            xticklabels=['Non-Flood', 'Flood'],
                            yticklabels=['Non-Flood', 'Flood'])
                plt.title(f"{name}\nF1: {data['metrics']['f1']:.3f}")
                plt.ylabel('True Label')
                plt.xlabel('Predicted Label')
            plt.tight_layout()
            plt.show()

    else:
        # Warning if paths are wrong or training skipped
        print("\n‚ùå TRAINING SKIPPED: Could not find dataset folders OR no experiments selected.")
        print(f"   Flood Dir detected: {flood_dir}")
        print(f"   Non-Flood Dir detected: {non_flood_dir}")
        print("   Please check your Drive paths or Zip extraction.")

    print("\n‚úÖ Training Script Complete.")
    print("‚¨áÔ∏è  SCROLL DOWN TO COPY THE STANDALONE INFERENCE CELL")


```

```python
# ==================================================================================
# ==================================================================================
#
#   7. STANDALONE INFERENCE CELL (COPY EVERYTHING BELOW TO A NEW COLAB CELL)
#      This section contains self-contained code to load a model and test images.
#
# ==================================================================================
# ==================================================================================

import os
import torch
import torch.nn as nn
from torchvision import transforms, models
from PIL import Image
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import cv2
import seaborn as sns
from sklearn.metrics import confusion_matrix, f1_score
from tqdm.auto import tqdm

# --- CONFIGURATION FOR STANDALONE TESTING ---
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# RESTORED ORIGINAL PATHS FROM YOUR REQUEST:
MODEL_LOAD_PATH = "model_efficientnet_baseline.pth"  # Defaulting to EfficientNet Baseline
TEST_IMAGE_PATH = "/content/drive/MyDrive/Scientific ML Project/New Final Project/Dataset/New photos/flood/48.jpg"

# UPDATED TEST PATHS:
TEST_FOLDER_FLOOD = "/content/drive/MyDrive/Scientific ML Project/New Final Project/Dataset/Alleyphoto_test/flood"
TEST_FOLDER_NON_FLOOD = "/content/drive/MyDrive/Scientific ML Project/New Final Project/Dataset/Alleyphoto_test/no_flood"

# --- MODEL DEFINITION (Must match training structure) ---
def get_model():
    # Load Pre-trained EfficientNet_B0
    model = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)
    # Modify the classifier head for 2 classes
    num_ftrs = model.classifier[1].in_features
    model.classifier[1] = nn.Linear(num_ftrs, 2)
    return model.to(DEVICE)

# --- UTILITIES ---
class GradCAM:
    """Updated for EfficientNet_B0 structure."""
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.handle_forward = target_layer.register_forward_hook(self.save_activation)
        self.handle_backward = target_layer.register_full_backward_hook(self.save_gradient)

    def save_activation(self, module, input, output):
        self.activation = output

    def save_gradient(self, module, grad_input, grad_output):
        self.gradients = grad_output[0]

    def remove_hooks(self):
        self.handle_forward.remove()
        self.handle_backward.remove()

    def __call__(self, x, class_idx=None):
        output = self.model(x)
        if class_idx is None:
            class_idx = torch.argmax(output, dim=1)
        self.model.zero_grad()
        target = output[0, class_idx]
        target.backward()

        gradients = self.gradients[0]
        pooled_gradients = torch.mean(gradients, dim=[1, 2])
        activations = self.activation[0]

        for i in range(activations.shape[0]):
            activations[i, :, :] *= pooled_gradients[i]

        heatmap = torch.mean(activations, dim=0).cpu().detach()
        heatmap = np.maximum(heatmap, 0)
        heatmap /= torch.max(heatmap) + 1e-8
        return heatmap.numpy()

def inverse_normalize(tensor):
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = tensor.cpu().numpy().transpose((1, 2, 0))
    img = std * img + mean
    img = np.clip(img, 0, 1)
    return img

def predict_single_image(model, image_path, device='cuda'):
    path_obj = Path(image_path)
    if not path_obj.exists():
        print(f"‚ùå Error: Image not found at {image_path}")
        return

    transform = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    try:
        img_pil = Image.open(path_obj).convert("RGB")
    except Exception as e:
        print(f"‚ùå Could not open image: {e}")
        return

    input_tensor = transform(img_pil).unsqueeze(0).to(device)
    model.eval()

    # Target EfficientNet's last feature block
    target_layer = model.features[-1]
    grad_cam = GradCAM(model, target_layer)

    output = model(input_tensor)
    probs = torch.nn.functional.softmax(output, dim=1)
    confidence, predicted_class = torch.max(probs, 1)

    confidence_score = confidence.item()
    pred_idx = predicted_class.item()
    class_names = {0: "Non-Flood", 1: "Flood"}
    pred_label = class_names.get(pred_idx, "Unknown")

    heatmap = grad_cam(input_tensor, class_idx=pred_idx)
    grad_cam.remove_hooks()

    img_np = inverse_normalize(input_tensor.squeeze(0))
    heatmap = cv2.resize(heatmap, (img_np.shape[1], img_np.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)

    superimposed = np.float32(heatmap) / 255 + img_np
    superimposed = superimposed / np.max(superimposed)

    print(f"\nüì∏ Image: {os.path.basename(image_path)}")
    print(f"ü§ñ Prediction: {pred_label}")
    print(f"üìä Confidence: {confidence_score:.4f} ({confidence_score*100:.2f}%)")

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img_np)
    plt.title(f"Prediction: {pred_label} ({confidence_score:.1%})")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(superimposed)
    plt.title("Grad-CAM Attention")
    plt.axis('off')
    plt.show()

def run_batch_test(model, folder_path, true_label_name, device=DEVICE):
    folder_path = Path(folder_path)
    if not folder_path.exists():
        print(f"‚ö†Ô∏è Folder not found or skipped: {folder_path}")
        return

    print(f"\nüìÇ Testing images in: {folder_path}")
    image_files = [f for f in folder_path.iterdir() if f.suffix.lower() in ('.jpg', '.jpeg', '.png')]
    total_count = len(image_files)

    if total_count == 0:
        print("   No images found.")
        return

    # Accumulate predictions for metrics
    all_preds = []
    all_labels = []
    # 1 for Flood, 0 for Non-Flood
    target_label = 1 if true_label_name == "Flood" else 0

    transform = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    print(f"   Found {total_count} images. Processing...")
    correct_count = 0

    for img_path in tqdm(image_files, leave=True, desc=f"Testing {true_label_name}"):
        try:
            img_pil = Image.open(img_path).convert("RGB")
            input_tensor = transform(img_pil).unsqueeze(0).to(device)

            model.eval()
            with torch.no_grad():
                output = model(input_tensor)
                probs = torch.nn.functional.softmax(output, dim=1)
                confidence, predicted_idx = torch.max(probs, 1)

            pred_idx = predicted_idx.item()
            all_preds.append(pred_idx)
            all_labels.append(target_label)

            if pred_idx == target_label:
                correct_count += 1

        except Exception as e:
            print(f"   Error processing {img_path.name}: {e}")

    # Results and Metrics
    accuracy = (correct_count / total_count) * 100
    f1 = f1_score(all_labels, all_preds, labels=[0, 1], average='binary', zero_division=0)
    cm = confusion_matrix(all_labels, all_preds, labels=[0, 1])

    print(f"   üëâ Results for {true_label_name} folder:")
    print(f"      Accuracy: {correct_count}/{total_count} ({accuracy:.2f}%)")
    print(f"      F1 Score: {f1:.4f}")

    # Plot Simple Confusion Matrix
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Non-Flood', 'Flood'],
                yticklabels=['Non-Flood', 'Flood'])
    plt.title(f"CM: {true_label_name} Test")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

# --- MAIN INFERENCE EXECUTION ---
# This block runs automatically if you run the whole script.
# It also serves as the entry point if you copy this section to a new cell.
if __name__ == "__main__":
    # Ensure Drive is mounted for paths to work (redundant check for safety)
    try:
        from google.colab import drive
        if not os.path.exists('/content/drive'):
            drive.mount('/content/drive', force_remount=True)
    except:
        pass

    if os.path.exists(MODEL_LOAD_PATH):
        print(f"üîÑ Loading checkpoint from {MODEL_LOAD_PATH}...")

        # Load the raw checkpoint
        # FIX: Added weights_only=False to handle dictionary containing numpy scalars/history
        checkpoint = torch.load(MODEL_LOAD_PATH, map_location=DEVICE, weights_only=False)

        model = get_model()

        # Check if it's the new dictionary format or old state_dict
        if isinstance(checkpoint, dict) and 'model_state_dict' in checkpoint:
            print("   ‚úÖ Full checkpoint dictionary detected!")
            model.load_state_dict(checkpoint['model_state_dict'])

            # Print Stored Metrics for Easy Access
            if 'test_metrics' in checkpoint:
                tm = checkpoint['test_metrics']
                print("\nüìä --- Stored Test Metrics (from training) ---")
                print(f"   Accuracy:  {tm.get('acc', 'N/A'):.4f}")
                print(f"   Precision: {tm.get('prec', 'N/A'):.4f}")
                print(f"   Recall:    {tm.get('rec', 'N/A'):.4f}")
                print(f"   F1 Score:  {tm.get('f1', 'N/A'):.4f}")

            if 'history' in checkpoint:
                print("\nüìâ --- Training History Available ---")
                hist = checkpoint['history']
                # Quick plot of history if available
                plt.figure(figsize=(10, 4))
                plt.subplot(1, 2, 1)
                plt.plot(hist['train_loss'], label='Train Loss')
                plt.plot(hist['val_loss'], label='Val Loss')
                plt.title("Loss Curve")
                plt.legend()

                plt.subplot(1, 2, 2)
                plt.plot(hist['val_f1'], label='Val F1', color='orange')
                plt.title("F1 Score Curve")
                plt.legend()
                plt.show()

        else:
            print("   ‚ö†Ô∏è Legacy state_dict detected (Weights only). Loading weights...")
            model.load_state_dict(checkpoint)

        # 1. Test Single Image
        if TEST_IMAGE_PATH:
            predict_single_image(model, TEST_IMAGE_PATH, device=DEVICE)

        # 2. Test Folders
        if TEST_FOLDER_FLOOD:
            run_batch_test(model, TEST_FOLDER_FLOOD, "Flood", device=DEVICE)

        if TEST_FOLDER_NON_FLOOD:
            run_batch_test(model, TEST_FOLDER_NON_FLOOD, "Non-Flood", device=DEVICE)

    else:
        print(f"‚ùå Model file '{MODEL_LOAD_PATH}' not found. Please run training first.")
```
